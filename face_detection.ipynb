```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
import face_recognition
import cv2
%matplotlib inline
```

```python
image = cv2.imread('C:\\Users\\SHIRISH B .RUPERA\\Pictures\\100NIKON\\DSCN0031.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
face_locations = face_recognition.face_locations(image)
no_of_faces= len(face_locations)
print("Found {} face in input image".format(no_of_faces))
```

```python
plt.imshow(image)
ax = plt.gca()
for face_location in face_locations:
    top,right,bottom,left = face_location
    x,y,w,h = left,top,right,bottom
    print("A face is located at pixe location top:{} ,left:{},bottom:{},right:{}".format(x,y,w,h))
    rect=Rectangle((x,y),w-x,h-y,fill=False,color='blue')
    ax.add_patch(rect)
plt.show()
```
