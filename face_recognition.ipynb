```python
image = cv2.imread("C:\\Users\\SHIRISH B .RUPERA\\Pictures\\101NIKON\\vat.jpg")
img1 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image = cv2.imread("C:\\Users\\SHIRISH B .RUPERA\\Pictures\\101NIKON\\dad.jpg")
img2 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image = cv2.imread("C:\\Users\\SHIRISH B .RUPERA\\Pictures\\101NIKON\\mom.jpg")
img3 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
img1_encoding = face_recognition.face_encodings(img1)[0]
img2_encoding = face_recognition.face_encodings(img2)[0]
img3_encoding = face_recognition.face_encodings(img3)[0]
known_face_encodings = [img1_encoding,img2_encoding,img3_encoding]
```

```python
image = cv2.imread("C:\\Users\\SHIRISH B .RUPERA\\Pictures\\101NIKON\\mom_dad_test.jpg")
unknown_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(unknown_img)
unknown_encodings=face_recognition.face_encodings(unknown_img)
```

```python
from scipy.spatial import distance
```

```python
for unknown_face_encoding in unknown_encodings:
    results=[]
    for known_face_encoding in known_face_encodings:
        d= distance.euclidean(known_face_encoding,unknown_face_encoding)
        results.append(d)
    threshold=0.6
    results = np.array(results) <= threshold
    if results[0]:
        name="img1"
    elif results[1]:
        name="img2"
    elif results[2]:
        name="img3"
    print(f"Found {name} in photo")
    
```
